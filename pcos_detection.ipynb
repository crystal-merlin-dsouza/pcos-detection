DATA CLEANING

```python
# Remove leading and trailing spaces from column names
data.columns = data.columns.str.strip()

# Verify the column names
print(data.columns)
```

```python
#working only with numerical data so we will convert
data['II    beta-HCG(mIU/mL)'] = pd.to_numeric(data['II    beta-HCG(mIU/mL)'],errors='coerce')
data['AMH(ng/mL)'] = pd.to_numeric(data['AMH(ng/mL)'],errors='coerce')
```

```python
data.isna()      #identify missing values(NaN)
```

```python
data.isna().sum()     #count missing values in each column
```

```python
# Drop non-relevant columns
df = data.drop(['Sl. No', 'Patient File No.', 'Blood Group','Unnamed: 44'],axis=1)
```

```python
# Fill missing values with the mean for numerical columns
df['Marraige Status (Yrs)'].fillna(df['Marraige Status (Yrs)'].mean(), inplace=True)
df['II    beta-HCG(mIU/mL)'].fillna(df['II    beta-HCG(mIU/mL)'].mean(), inplace=True)
df['AMH(ng/mL)'].fillna(df['AMH(ng/mL)'].mean(), inplace=True)

# Fill missing values with mode for categorical columns
df['Fast food (Y/N)'].fillna(df['Fast food (Y/N)'].mode()[0], inplace=True)
```

MODEL BUILDING

```python
# Feature Selection
# Selecting the target variable
y = df['PCOS (Y/N)']

# Selecting the recommended features
X = df[['BMI',
         'Age (yrs)',
         'Weight (Kg)',
         'Cycle length(days)',
         'FSH(mIU/mL)',
         'LH(mIU/mL)',
         'AMH(ng/mL)',
         'TSH (mIU/L)',
         'PRL(ng/mL)',
         'Weight gain(Y/N)',
         'hair growth(Y/N)',
         'Skin darkening (Y/N)',
         'Pimples(Y/N)',
         'Hair loss(Y/N)',
         'Endometrium (mm)',
         'Follicle No. (L)',
         'Follicle No. (R)']]
```

```python
#splitting dataset into training and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

#training the logistic regression model on the training set
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train,y_train)
```

```python
#Make Predictions
y_pred = model.predict(X_test)
```

```python
#Evaluate the Model

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

accuracy = accuracy_score(y_test,y_pred)
conf_matrix = confusion_matrix(y_test,y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
print("\nConfusion Matrix:")
print(conf_matrix)
print("\n")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No PCOS', 'PCOS'], yticklabels=['No PCOS', 'PCOS'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("\nClassification Report:")
print(classification_rep)

print(f"Model Accuracy: {accuracy * 100:.2f}%")
```
